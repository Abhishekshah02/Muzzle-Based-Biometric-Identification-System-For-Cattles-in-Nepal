Copying all the iamges from all the sub-folder and pasting to teh unlabeled folder

```python
import os
import shutil
from pathlib import Path

# === Set your source and destination folders ===
source_root = r"D:\Abhishek\INDIVIDUAL SUBJECTS Data"        # Folder with subfolders
destination_folder = r"D:\Abhishek\dataset\muzzle-detection\unlabeled"  # Target folder (should be empty)

# Create destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Define image file extensions
image_extensions = {".jpg", ".jpeg", ".png"}

# Traverse and copy
for root, _, files in os.walk(source_root):
    for file in files:
        if Path(file).suffix.lower() in image_extensions:
            source_path = os.path.join(root, file)
            destination_path = os.path.join(destination_folder, file)

            shutil.copy2(source_path, destination_path)

print("✅ All images copied successfully.")
```

```python
import os
import shutil
from ultralytics import YOLO
from pathlib import Path
import cv2
import torch

# === USER CONFIG ===
model_path = r"D:\Abhishek\muzzle_project\yolov8_muzzle16\weights\best.pt"
image_folder = r"D:\Abhishek\dataset\muzzle-detection\unlabeled"  # input images
output_images = r"D:\Abhishek\dataset\muzzle-detection\labeled\images"
output_labels = r"D:\Abhishek\dataset\muzzle-detection\labeled\labels"
confidence_threshold = 0.7
# =====================

# Make output folders
os.makedirs(output_images, exist_ok=True)
os.makedirs(output_labels, exist_ok=True)

# Load model
model = YOLO(model_path)

# Run prediction on all images
# results = model.predict(source=image_folder, conf=confidence_threshold, save_txt=False, imgsz=640)
import glob

image_paths = glob.glob(os.path.join(image_folder, "*.jpg"))  # adjust if using png
results = []

for image_path in image_paths:
    try:
        result = model.predict(
            source=image_path,
            conf=confidence_threshold,
            save_txt=False,
            imgsz=640,
            device='cuda'  # or 'cpu'
        )
        results.extend(result)
    except Exception as e:
        print(f"❌ Skipped {image_path} due to error: {e}")


saved = 0

for i, r in enumerate(results):
    # If no detections above threshold, skip
    if len(r.boxes) == 0:
        continue

    image_name = Path(r.path).name
    label_name = Path(r.path).stem + ".txt"

    # Check if any detection has confidence ≥ 0.7
    high_conf_boxes = [b for b in r.boxes.conf if b.item() >= confidence_threshold]
    if len(high_conf_boxes) == 0:
        continue

    # Save image with boxes
    img_with_boxes = r.plot()
    out_img_path = os.path.join(output_images, image_name)
    cv2.imwrite(out_img_path, img_with_boxes)

    # Save label file (YOLO format)
    out_label_path = os.path.join(output_labels, label_name)
    with open(out_label_path, 'w') as f:
        for box in r.boxes:
            conf = box.conf.item()
            if conf < confidence_threshold:
                continue
            cls_id = int(box.cls.item())
            x_center, y_center, width, height = box.xywhn[0].tolist()
            f.write(f"{cls_id} {x_center} {y_center} {width} {height}\n")

    saved += 1

print(f"✅ Saved {saved} images and labels with confidence ≥ {confidence_threshold * 100:.0f}%")
```
